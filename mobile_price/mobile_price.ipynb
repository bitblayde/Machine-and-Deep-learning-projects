Self-normalization network:

```python
class DeepSequentialNetwork(keras.models.Model):
    def __init__(self, input_layers, output_dim, units = [], **kwargs):
        super().__init__(**kwargs)
        self.module = keras.models.Sequential()

        for i in range(input_layers):
            self.module.add(keras.layers.Dense(units[i], kernel_initializer="lecun_normal"))
            self.module.add(keras.layers.BatchNormalization())
            self.module.add(keras.layers.Activation("selu"))
            
        self.output_layer = keras.layers.Dense(output_dim)
    
    def call(self, inputs):
        x = self.module(inputs)
        x = self.output_layer(x)
        x = tf.nn.softmax(x)
        return x
```

```python
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

custom_network1 = DeepSequentialNetwork(5, 4, [2000, 1000, 500, 250, 100])
train_and_compile(custom_network1, X_train_normalized, y_train, lr = 1e-3)
```

```python
class Dense_Layer(keras.layers.Layer):
    def __init__(self, units, activation_function, noise = False, **kwargs):
        super().__init__(**kwargs)
        self.units = units
        self.activation_function = keras.activations.get(activation_function)
        self.noise = noise
        
    def build(self, batch_input_shape):
        self.k = self.add_weight(shape = (batch_input_shape[-1], self.units), initializer="glorot_normal")
        self.b = self.add_weight(shape = (self.units,), initializer="zeros")
        super().build(batch_input_shape)
    
    def call(self, inputs, training = None):
        X = inputs @ self.k + self.b
        
        if not self.noise:
            X = self.activation_function(X)
        elif self.noise and training:
            X = tf.random.normal(tf.shape(X), stddev = self.stddev)
        
        return X

    def compute_output_shape(self, batch_input_shape):
        return tf.TensorShape(batch_input_shape.as_list()[:-1] + [self.units])
```

```python
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

inputs = keras.Input(shape=(X_train_normalized.shape[-1], ) )

layer1 = Dense_Layer(2000, "selu", None, input_shape=X_train_normalized.shape[1:])
layer2 = Dense_Layer(500, "selu", None)
layer3 = Dense_Layer(100, "selu", None)

X = layer1(inputs)
X = keras.layers.Dropout(0.3)(X)
X = layer2(X)
X = keras.layers.Dropout(0.3)(X)
X = layer3(X)
X = keras.layers.Dropout(0.3)(X)

output = keras.layers.Dense(4, activation = "softmax")(X)

custom_network2 = keras.Model(inputs=inputs, outputs=output, name="custom_network2")

train_and_compile(custom_network2, X_train_normalized, y_train, lr=1e-3)
```

```python
custom_network2.evaluate(X_val_normalized, y_val)
```

```python
class ResBlock(keras.layers.Layer):
    def __init__(self, layers, units = [], **kwargs):
        super().__init__(**kwargs)
        self.backbone = []
        
        for i in range(layers):
            self.backbone.append(keras.layers.Dense(units[i], activation = "selu", kernel_initializer = "lecun_normal"))
            
    def call(self, X):
        inputs = X
        for i in range(len(self.backbone)):
            X = self.backbone[i](X)
            
        return inputs + X
```

```python
class ResidualBlock(keras.models.Model):
    def __init__(self, output_dim = 4, **kwargs):
        super().__init__(**kwargs)
        self.dense1 = keras.layers.Dense(1000, activation = "selu", kernel_initializer = "lecun_normal")
        self.resblock1 = ResBlock(2, [1000, 1000])
        self.resblock2 = ResBlock(2, [100, 1000])
        self.dense2 = keras.layers.Dense(output_dim, activation = "softmax", kernel_initializer = "lecun_normal")
        
    def call(self, X):
        X = self.dense1(X)
        
        for i in range(4):
            X = self.resblock1(X)
        
        X = self.resblock2(X)
        X = self.dense2(X)
        
        return X
```
